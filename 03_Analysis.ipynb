### Imports

```python
import pandas as pd

from sklearn.feature_selection import mutual_info_classif
```

### Constants

```python
possible_outputs = ["Hypertension", "Heart Disease", "Stroke", "Diabetes"]
```

All Possible Features:

- Age
- Sex
- High Blood Pressure
- Smoking
- High Cholesterol
- BMI
- Physical Activity
- Fasting Blood Sugar
- Chest Pain Type
- Cholesterol
- Resting Blood Pressure
- (o) Diabetes
- (o) Heart Disease
- (o) Stroke
- (o) Hypertension

```python
def filter_df(df, target_column):
    # Delete all of the rows that have the output -1
    filtered_df = df[df[target_column] != -1]

    # Delete all of the columns that are all -1
    filtered_df = filtered_df.loc[:, (filtered_df != -1).any(axis=0)]

    X = filtered_df.drop(columns=[target_column])
    y = filtered_df[target_column]
    return X, y
```

```python
def compute_mutual_information(X, y):
    # Compute mutual information
    mutual_info = mutual_info_classif(X, y, random_state=42)
    
    feature_importances = pd.DataFrame({
        'Feature': X.columns,
        'Mutual Information': mutual_info
    })
    
    return feature_importances.sort_values(by='Mutual Information', ascending=False)
```

### Read the data

```python
df = pd.read_csv('./combined_dataset.csv')
```

```python
for output in possible_outputs:
    print(f"Mutual information for {output}.")
    X, y = filter_df(df, output)
    print(compute_mutual_information(X, y))
```

```python
for output in possible_outputs:
    print(f"Mutual information for {output}.")
    X, y = filter_df(df, output)
    X.to_csv(f'./{output}_filtered.csv', index = False)
```

### Mutual Information

```python
import pandas as pd
import matplotlib.pyplot as plt

# Mutual information data
data = {
    "Hypertension": {
        "Feature": ["Age", "High Cholesterol", "Fasting Blood Sugar", "Diabetes", "Smoking", 
                    "Physical Activity", "Chest Pain Type", "BMI", "Sex", "Heart Disease", "Stroke"],
        "Mutual Information": [0.076111, 0.074496, 0.070804, 0.064733, 0.053526, 
                               0.052989, 0.050610, 0.030664, 0.023358, 0.021102, 0.008118]
    },
    "Heart Disease": {
        "Feature": ["Chest Pain Type", "Fasting Blood Sugar", "Hypertension", "Smoking", 
                    "Sex", "Physical Activity", "Age", "High Cholesterol", "Diabetes", "Stroke", "BMI"],
        "Mutual Information": [0.083462, 0.071734, 0.041849, 0.038706, 0.037447, 
                               0.036785, 0.033951, 0.027101, 0.017372, 0.012593, 0.007821]
    },
    "Stroke": {
        "Feature": ["Fasting Blood Sugar", "Smoking", "Physical Activity", "Sex", "Hypertension", 
                    "High Cholesterol", "Heart Disease", "Age", "Diabetes", "BMI"],
        "Mutual Information": [0.075037, 0.040537, 0.039841, 0.039655, 0.033732, 
                               0.017607, 0.012526, 0.012306, 0.010286, 0.005242]
    },
    "Diabetes": {
        "Feature": ["Smoking", "Hypertension", "Physical Activity", "Fasting Blood Sugar", 
                    "High Cholesterol", "BMI", "Sex", "Age", "Heart Disease", "Stroke"],
        "Mutual Information": [0.055653, 0.054665, 0.049404, 0.046001, 0.041131, 
                               0.036363, 0.028969, 0.024357, 0.014658, 0.005696]
    }
}

# Creating subplots for each condition
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Mutual Information for Hypertension, Heart Disease, Stroke, and Diabetes", fontsize=16)

# Plot for each condition
conditions = ['Hypertension', 'Heart Disease', 'Stroke', 'Diabetes']
axes = axs.ravel()

for i, condition in enumerate(conditions):
    df = pd.DataFrame(data[condition])
    df.plot.bar(x='Feature', y='Mutual Information', ax=axes[i], legend=False, color='#32a856')
    axes[i].set_title(f"Mutual Information for {condition}")
    axes[i].set_title(f"{condition}")
    axes[i].set_ylabel('Mutual Information')
    axes[i].set_xlabel('Feature')
    axes[i].grid(True)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for accuracies
conditions = ['Hypertension', 'Heart Disease', 'Stroke', 'Diabetes']
classifiers = ['Logistic Regression', 'Random Forest', 'SVC', 'Neural Network']

accuracies = {
    'Hypertension': [74.01, 71.90, 74.20, 74.40],
    'Heart Disease': [91.71, 90.47, 91.83, 91.88],
    'Stroke': [96.48, 95.75, 96.48, 96.48],
    'Diabetes': [87.06, 85.53, 87.03, 87.35]
}

# Plotting parameters
bar_width = 0.2
index = np.arange(len(conditions))

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 7))

# Create bars for each classifier
for i, classifier in enumerate(classifiers):
    accuracies_for_classifier = [accuracies[condition][i] for condition in conditions]
    ax.bar(index + i * bar_width, accuracies_for_classifier, bar_width, label=classifier)

# Add labels, title, and grid
ax.set_xlabel('Conditions')
ax.set_ylabel('Accuracy (%)')
ax.set_title('Accuracy of Classifiers by Condition')
ax.set_xticks(index + bar_width * 1.5)
ax.set_xticklabels(conditions)
ax.set_ylim(50, 100)
ax.legend()
ax.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```

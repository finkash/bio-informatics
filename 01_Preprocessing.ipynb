```python
import pandas as pd

from random import randint
```

### Utility functions

```python
def read_and_select_columns(file_path, selected_columns):
    data = pd.read_csv(file_path)
    
    # Check if the selected columns exist in the file
    missing_columns = [col for col in selected_columns if col not in data.columns]
    if missing_columns:
        raise ValueError(f"Missing columns in CSV file: {missing_columns}")
    
    return data[selected_columns]
```

```python
files = [
    "heart_failure_clinical_records_dataset.csv",
    "diabetes_binary_health_indicators_BRFSS2015.csv",
    "train_strokes.csv",
    "processed_cleveland_data.csv",
    "processed_hungarian_data.csv",
    "processed_switzerland_data.csv",
    "processed_va_data.csv",
    "MGE303_Term Project_Classification_Data (Disease).csv"
]

select_columns = [
    ["age", "sex", "high_blood_pressure", "diabetes", "smoking"],
    ["Age", "Sex", "HighChol", "HeartDiseaseorAttack", "HighBP", "BMI", "Stroke", "Diabetes_binary", "PhysActivity"],
    ["age", "gender", "avg_glucose_level", "heart_disease", "hypertension", "bmi", "stroke", "smoking_status"],
    ["age", "sex", "cp", "trestbps", "chol", "fbs", "num"],
    ["age", "sex", "cp", "trestbps", "chol", "fbs", "num"],
    ["age", "sex", "cp", "trestbps", "chol", "fbs", "num"],
    ["age", "sex", "cp", "trestbps", "chol", "fbs", "num"],
    ["AGE", "gender", "BP_HIGH", "BP_LWST", "TOT_CHOLE", "BLDS", "H_HDISE_YN", "HY_YN", "H_APOP_YN", "H_DIABML_YN", "DSQTY", "EXER_M"]
]
```

```python
filtered_datasets = [read_and_select_columns(f'./datasets/{files[i]}', select_columns[i]) for i in range(len(files))]
```

#### Process `processed_x_data.csv`

```python
filtered_datasets[3]
```

```python
def process_x_datasets():
    results = []
    for i in range(3, 3 + 4):
        results.append({
            "Age": filtered_datasets[i]["age"],
            "Sex": filtered_datasets[i]["sex"],
            "Chest Pain Type": filtered_datasets[i]["cp"],
            "Hypertension": filtered_datasets[i]["trestbps"].map(lambda x : -1 if x == "?" else (1 if int(x) >= 140 else 0)),
            "High Cholesterol": filtered_datasets[i]["chol"].map(lambda x : -1 if x == "?" else (1 if int(x) >= 220 else 0)),
            "Fasting Blood Sugar": filtered_datasets[i]["fbs"].map(lambda x : -1 if x == "?" else int(x)),
            "Heart Disease": filtered_datasets[i]["num"].map(lambda x : 1 if int(x) > 1 else 0)
        })

    return results
```

#### Process `MGE303_Term Project_Classification_Data (Disease).csv`

```python
filtered_datasets[7]
```

```python
p4 = {
    "Age": filtered_datasets[7]["AGE"],
    "Sex": filtered_datasets[7]["gender"],
    "Fasting Blood Sugar": filtered_datasets[7]["BLDS"].map(lambda x : 1 if x >= 126 else 0),
    "High Cholesterol": filtered_datasets[7]["TOT_CHOLE"].map(lambda x : 1 if x >= 220 else 0),
    "Heart Disease": filtered_datasets[7]["H_HDISE_YN"],
    "Hypertension": filtered_datasets[7]["HY_YN"],
    "Stroke": filtered_datasets[7]["H_APOP_YN"],
    "Diabetes": filtered_datasets[7]["H_DIABML_YN"],
    "Smoking": filtered_datasets[7]["DSQTY"].map(lambda x : 1 if x > 0 else 0),
    "Physical Activity": filtered_datasets[7]["EXER_M"].map(lambda x : 0 if x == "Never" or x == "Once a week" else 1)
}
```

### Collect all dictionaries in one place

```python
c = process_x_datasets()
dicts = [p1, p2, p3, c[0], c[1], c[2], c[3], p4]
dicts = [pd.DataFrame(x) for x in dicts]
```

```python
dicts[7]
```

### Combine all of the datasets into one big dataset

```python
combined_df = pd.concat(dicts, axis=0, ignore_index=True).fillna(-1)

# Convert all float columns to ints
for col in combined_df.columns:
    if combined_df[col].dtype == 'float64':
        combined_df[col] = combined_df[col].astype('int')
```
